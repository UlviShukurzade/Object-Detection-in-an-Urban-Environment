## Write a function to display an image and the bounding boxes

Implement the `display_images` function below. This function takes a batch as an input and display an image with its corresponding bounding boxes. The only requirement is that the classes should be color coded (eg, vehicles in red, pedestrians in blue, cyclist in green).

```python
def display_images(batch):
    """
    input - it takes batches from dataset
    output - plots the frames from dataset with corresponting annotaions

    """
    # ADD CODE HEREs
    colors = {
        1:"red", 
        2:"blue", 
        4:"green"
    }
    
    #shape of image
    w, h, c = batch["image"].shape
    
    
    fig, ax = plt.subplots(figsize=(7,7))
    
    #converting data into visualisable form
    ax.imshow(batch["image"].numpy().astype("uint8"))
    
    groundtruth_classes = batch["groundtruth_classes"].numpy()
    
    
    for i in range(len(batch["groundtruth_boxes"])):
        ymin, xmin, ymax, xmax = batch["groundtruth_boxes"][i]
        ground_class = groundtruth_classes[i]

        rect = patches.Rectangle((xmin*w, ymin*h), (xmax-xmin)*w, (ymax-ymin)*h, 
                                            edgecolor=colors[ground_class], facecolor="none")
        ax.add_patch(rect)
        
    plt.show()
```

## Display 10 images 

Using the dataset created in the second cell and the function you just coded, display 10 random images with the associated bounding boxes. You can use the methods `take` and `shuffle` on the dataset.

```python
# Display 10 random images in dataset

for batch in dataset.take(10):
    display_images(batch)
```

```python
class_counts = {1: 0, 2: 0, 4: 0}
for batch in dataset.take(5000):
    for gt_cl in batch["groundtruth_classes"].numpy():
        class_counts[gt_cl] += 1

class_counts        

```
